```python
import numpy as np
import matplotlib.pyplot as plt

# Função para calcular a probabilidade de Poisson
def poisson(k, lamb):
    return (lamb ** k) * np.exp(-lamb) / np.math.factorial(k)

# Média de falhas a cada duas semanas
media_falhas = 6
# Média de falhas por semana
lamb = media_falhas / 2

# Probabilidade de zero falha em uma semana
P_X_0 = poisson(0, lamb)
# Probabilidade de uma falha em uma semana
P_X_1 = poisson(1, lamb)
# Probabilidade de duas falhas em uma semana
P_X_2 = poisson(2, lamb)

# Probabilidade de pelo menos duas falhas em uma semana
P_X_ge_2 = 1 - (P_X_0 + P_X_1 + P_X_2)

print("Probabilidade de pelo menos 2 falhas em uma semana:", P_X_ge_2)

# Histograma da variável analisada
semana = np.arange(0, 15)
probabilidade = [poisson(k, lamb) for k in semana]

plt.bar(semana, probabilidade, color='blue', alpha=0.7)
plt.xlabel('Número de Falhas')
plt.ylabel('Probabilidade')
plt.title('Histograma da Distribuição de Poisson')
plt.show()
```
