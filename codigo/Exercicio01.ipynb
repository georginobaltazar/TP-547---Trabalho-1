```python
import matplotlib.pyplot as plt
import numpy as np

def gerador_aleatorio(x0, n):
    sequencia = [x0]
    for i in range(n - 1):
        xi = (5 * sequencia[-1]) % 7
        sequencia.append(xi)
    return sequencia

# Simulação com semente x0 = 4
print("Semente x0 = 4:")
sequencia_x0_4 = gerador_aleatorio(4, 10)
print(sequencia_x0_4)

# Simulação com semente x0 = 7
print("Semente x0 = 7:")
sequencia_x0_7 = gerador_aleatorio(7, 10)
print(sequencia_x0_7)

# Plotagem dos gráficos
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(np.arange(0, len(sequencia_x0_4)), sequencia_x0_4, marker='o', color='b')
plt.title('Sequência gerada com semente x0 = 4')
plt.xlabel('Iteração')
plt.ylabel('Valor')
plt.grid(True)
plt.xticks(np.arange(0, len(sequencia_x0_4), step=1))

plt.subplot(2, 1, 2)
plt.plot(np.arange(0, len(sequencia_x0_7)), sequencia_x0_7, marker='o', color='r')
plt.title('Sequência gerada com semente x0 = 7')
plt.xlabel('Iteração')
plt.ylabel('Valor')
plt.grid(True)
plt.xticks(np.arange(0, len(sequencia_x0_7), step=1))

plt.tight_layout()
plt.show()
```
