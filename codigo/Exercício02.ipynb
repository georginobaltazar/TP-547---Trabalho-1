```python
import math

# Função para calcular a probabilidade de Poisson
def poisson_probability(k, lamb):
    return (lamb**k * math.exp(-lamb)) / math.factorial(k)

# Parâmetros do problema
lamb = 6  # média de chamadas por hora

# a. Probabilidade de não receber chamadas em uma determinada hora (k = 0)
prob_a = poisson_probability(0, lamb)
print("a. Probabilidade de não receber chamadas em uma hora:", prob_a)

# b. Probabilidade de receber menos de oito chamadas em uma determinada hora (soma de k de 0 a 7)
prob_b = sum(poisson_probability(k, lamb) for k in range(8))
print("b. Probabilidade de receber menos de oito chamadas em uma hora:", prob_b)

# c. Número médio de chamadas por hora
media = lamb
print("c. Número médio de chamadas por hora:", media)

# d. Variância de C
variancia = lamb
print("d. Variância de C:", variancia)

# e. Desvio padrão de C
desvio_padrao = math.sqrt(variancia)
print("e. Desvio padrão de C:", desvio_padrao)
```
