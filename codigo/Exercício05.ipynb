```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da distribuição exponencial
tempo_medio = 28
lambda_ = 1 / tempo_medio

# Definindo o intervalo de valores de x
x = np.linspace(0, 100, 1000)  # Valores de 0 a 100 dias

# Calculando a PDF da distribuição exponencial específica para a variável analisada
pdf = lambda_ * np.exp(-lambda_ * x)

# Traçando a PDF
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, color='blue', label='PDF da variável analisada')
plt.title('Função Densidade de Probabilidade (PDF) da Variável Analisada')
plt.xlabel('Número de dias de antecedência')
plt.ylabel('Densidade de probabilidade')
plt.legend()
plt.grid(True)
plt.show()
```
