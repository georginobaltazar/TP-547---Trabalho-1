```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da distribuição binomial
n = 8  # número de pistões no lote
p = 0.15  # probabilidade de um pistão ser rejeitado

# Simulação da distribuição binomial
num_simulations = 10000
rejeitados_simulacao = np.random.binomial(n, p, num_simulations)

# Contagem dos resultados da simulação
unique, counts = np.unique(rejeitados_simulacao, return_counts=True)
probabilidades_simuladas = counts / num_simulations

# Plotagem do histograma
plt.bar(unique, probabilidades_simuladas, color='blue', alpha=0.7)
plt.title('Histograma da distribuição de pistões rejeitados')
plt.xlabel('Número de pistões rejeitados')
plt.ylabel('Probabilidade')
plt.xticks(np.arange(0, n+1, 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
