```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfc

# Parâmetros
EbN0dB = np.arange(0, 22, 1)
EbN0lin = 10**(EbN0dB / 10.0)

# Cálculo do BER para cada esquema de modulação
ber_16qam = (4/np.log2(16)) * (1 - 1/np.sqrt(16)) * erfc(np.sqrt(3 * EbN0lin * np.log2(16) / (2 * (16-1))))
ber_32qam = (4/np.log2(32)) * (1 - 1/np.sqrt(32)) * erfc(np.sqrt(3 * EbN0lin * np.log2(32) / (2 * (32-1))))
ber_64qam = (4/np.log2(64)) * (1 - 1/np.sqrt(64)) * erfc(np.sqrt(3 * EbN0lin * np.log2(64) / (2 * (64-1))))
ber_128qam = (4/np.log2(128)) * (1 - 1/np.sqrt(128)) * erfc(np.sqrt(3 * EbN0lin * np.log2(128) / (2 * (128-1))))


# Plotagem com ajustes
plt.figure(figsize=(8, 6))
plt.plot(EbN0dB, ber_16qam, marker='*', linestyle='-', color='blue', label='16-QAM', markerfacecolor='none')
plt.plot(EbN0dB, ber_32qam, marker='o', linestyle='-', color='red', label='32-QAM', markerfacecolor='none')
plt.plot(EbN0dB, ber_64qam, marker='^', linestyle='-', color='green', label='64-QAM', markerfacecolor='none')
plt.plot(EbN0dB, ber_128qam, marker='s', linestyle='-', color='black', label='128-QAM', markerfacecolor='none')

plt.xlabel('Eb/N0 (dB)')
plt.ylabel('BER')
plt.title('Simulated BER performance of M-QAM for M=16, 32, 64, 128 over AWGN channel')
plt.yscale('log')
plt.xlim(0, 21)
plt.ylim(1e-6, 1)
plt.xticks(np.arange(0, 22, 3))
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
