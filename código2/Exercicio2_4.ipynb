## Método de Monte Carlo

a)

```python
import numpy as np
import matplotlib.pyplot as plt

# Número de amostras para o método de Monte Carlo
N = 100000

# Gerando N números aleatórios uniformemente distribuídos no intervalo [0, 1]
x = np.random.uniform(0, 1, N)

# Calculando a função para as amostras
y = ((1 - x ** 2) ** 1.5)

# Calculando a média dos valores
integral = np.mean(y)

# Imprime o valor da integral
print("O Valor da integral a) é:", integral)

# Criando os valores do eixo x para plotagem
x_values = np.linspace(0, 1, 100)

# Plotando a função
plt.plot(x_values, ((1 - (x_values ** 2)) ** 1.5))

# Preenchendo a área sob a curva
plt.fill_between(x_values, ((1 - (x_values ** 2)) ** 1.5), color='lightblue')

# Adicionando o rótulo com o valor da área
plt.text(0.2, 0.4, f'Área = {integral:.3f}', fontsize=14)

# Adicionando rótulos e título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Estimativa da Área: ((1-(x**2))**1.5)')

# Mostrando o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Número de amostras para o método de Monte Carlo
N = 1000000

# Gerando N números aleatórios uniformemente distribuídos no intervalo [0, 1]
y = np.random.uniform(0, 1, N)

# Transformando as amostras para o intervalo [1, +∞)
x = 1 / y - 1

# Calculando a função para as amostras
z = x * ((1 + x ** 2) ** (-2)) / y ** 2

# Calculando a média dos valores
integral = np.sum(z) / N

# Imprime o valor da integral
print("Valor da integral c):", integral)

# Criando os valores do eixo x para plotagem
x_values = np.linspace(0, 30, 500)

# Plotando a função
plt.plot(x_values, x_values * ((1 + x_values ** 2) ** (-2)))

# Preenchendo a área sob a curva
plt.fill_between(x_values, x_values * ((1 + x_values ** 2) ** (-2)), color='lightblue')

# Adicionando o rótulo com o valor da área
plt.text(5, 0.3, f'Área = {integral:.3f}', fontsize=14)

# Adicionando rótulos e título
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Estimativa de Área: x*((1 + x**2)**(-2))')

# Mostrando o gráfico
plt.show()
```

## Método por Importância

a)

```python
import numpy as np
import matplotlib.pyplot as plt

# Número de amostras para o método de Monte Carlo
N = 100000

# Gerar N amostras uniformemente distribuídas no intervalo [0, 1]
y = np.random.uniform(0, 1, N)

# Gerar as amostras de g(x)
ygx = y ** 0.4

# Calcular a integral pelo método de Monte Carlo
integral = np.sum((((1 - (ygx ** 2)) ** 1.5)) / ((5 / 2) * ygx ** 1.5)) / N

# Imprimir o valor da integral
print(f'O valor da Integral a) por importância é = {integral:.4f}')

# Plotar o gráfico
x = np.linspace(0, 1, N)
curve_y = (((1 - (x ** 2)) ** 1.5))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='lightblue')

# Adicionar texto com o valor da área
plt.text(0.5, 0.8, f'Área = {integral:.4f}', fontsize=14)

# Definir limites dos eixos x e y
plt.xlim(-0.01, 1.1)
plt.ylim(-0.01, 1.1)

# Adicionar rótulos aos eixos
plt.xlabel('x')
plt.ylabel('y')

# Adicionar título ao gráfico com a equação correta
plt.title(r'Estimativa de Área: $\frac{{(1 - x^2)^{3/2}}}{{\sqrt[5]{{x}}}}$')

# Mostrar o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Número de amostras para o método de Monte Carlo
N = 1000000

# Gerar N amostras uniformes no intervalo [0, 1]
y = np.random.uniform(0, 1, N)

# Calcular A para normalizar as amostras
A = 1 / (np.exp(3) - np.exp(-1))

# Aplicar a transformação para gerar as amostras de g(x)
ygx = np.log(y / A + (np.exp(-1))) - 1

# Calcular a integral pelo método de Monte Carlo
integral = np.sum((np.exp(ygx + ygx ** 2)) / (A * np.exp(ygx + 1))) / N

# Imprimir o valor da integral
print(f'O valor da Integral b) por importância é = {integral:.4f}')

# Plotar o gráfico
x = np.linspace(-2, 2, N)
curve_y = np.exp((x + x ** 2))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y, color='lightblue')

# Adicionar texto com o valor da área
plt.text(-1, 200, f'Área = {integral:.4f}', fontsize=14)

# Definir rótulos dos eixos
plt.xlabel('x')
plt.ylabel('y')

# Adicionar título ao gráfico com a equação correta e na cor vermelha
plt.title(r'Estimativa de Área: $e^{(x + x^2)}$')

# Mostrar o gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000000
# Gerar N amostras uniformes
y = np.random.uniform(0, 1, N)

# Definir A
A = 3

# Aplicar a transformação para gerar as amostras de g(x)
ygx = (((3 * y) / A)) ** (1 / 3)

# Calcular a integral pelo método de Monte Carlo
fx = (((1 / ygx - 1) * (1 + (((1 / ygx - 1)) ** 2)) ** (-2))) / (ygx ** 2)
gx = A * (ygx ** 2)
integral = np.sum(fx / gx) / N

# Imprimir o valor da integral
print(f'Ovalor da Integral c) por Importância é = {integral:.4f}')

# Plotar o gráfico
x = np.linspace(0, 30, 500)
curve_y = x * ((1 + x ** 2) ** (-2))
plt.plot(x, curve_y, color='black')
plt.fill_between(x, curve_y,color='lightblue')

# Adicionar texto com o valor da área
plt.text(5, 0.3, f'Área = {integral:.3f}', fontsize=14)

# Definir rótulos dos eixos
plt.xlabel('x')
plt.ylabel('y')

# Adicionar título ao gráfico com a equação correta
plt.title(r'Estimativa de Área: $\frac{{(1 / x - 1) \left(1 + \left(\frac{{1}}{{x}} - 1\right)^2\right)^{-2}}}{{x^2}}$')

# Mostrar o gráfico
plt.show()
```
