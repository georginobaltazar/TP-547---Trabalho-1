# Exercício 4

Este programa usa o Método de Monte Carlo para Estimar as seguintes Integrais:

image.png

```python
import numpy as np

# Número de amostras para o método de Monte Carlo
N = 100000

# Gerando N números aleatórios uniformemente distribuídos no intervalo [0, 1]
x = np.random.uniform(0, 1, N)

# Calculando y = sqrt(x)
y = np.sqrt(x)

# Cálculo da integral a), b) e c) pelo método de Monte Carlo (Intervalo de Importância)

# Calculando a integral 4)a
# Integrando a função (1 - y^2)^(3/2) / (2 * y) com respeito a y
# Utilizando o método de Monte Carlo para estimar a integral
integral1 = sum(pow((1 - pow(y, 2)), (3/2)) / (2 * y)) / N
print("A integral de 4)a é", integral1)

# Calculando a integral 4)b
# Integrando a função exp(y^2 + y) / (exp(y) / (exp(2) - exp(-2))) com respeito a y
# Utilizando o método de Monte Carlo para estimar a integral
yexp = np.log(np.exp(-2) + (np.exp(2) - np.exp(-2)) * x)
integral2 = sum((np.exp((yexp**2) + yexp)) / ((np.exp(yexp)) / (np.exp(2) - np.exp(-2)))) / N
print("A integral de 4)b é", integral2)

# Calculando a integral 4)c
# Integrando a função (1 / x^2) * ((1 / x) - 1) * (1 + ((1 / x) - 1)^2)^(-2) com respeito a x
# Utilizando o método de Monte Carlo para estimar a integral
integral3 = sum((1 / (x**2)) * ((1 / x) - 1) * (1 + ((1 / x) - 1)**2)**(-2)) / N
print("A integral de 4)c é", integral3)
```
