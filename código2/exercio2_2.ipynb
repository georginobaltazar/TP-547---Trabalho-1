# Este programa simula o lançamento de uma quantidade especificada de dados várias vezes e conta quantas vezes pelo menos um dos dados resultou em seis. Em seguida, calcula a probabilidade dividindo o número de experimentos bem-sucedidos pelo número total de experimentos.


```python
import numpy as np
import matplotlib.pyplot as plt

N = 100000  # Número de amostras
contador_seis = 0  # Contador para contabilizar o número de vezes que obtemos um seis
probabilidades = []  # Lista para armazenar as probabilidades ao longo das simulações

for _ in range(N):  # Realiza N simulações
    resultados = np.random.randint(1, 7, size=5)  # Simula o lançamento de 5 dados
    if 6 in resultados:  # Verifica se há pelo menos um seis nos resultados
        contador_seis += 1  # Incrementa o contador se houver pelo menos um seis
    probabilidade = contador_seis / (_ + 1)  # Calcula a probabilidade atual
    probabilidades.append(probabilidade)  # Adiciona a probabilidade à lista

# Calcula a probabilidade final
probabilidade_final = contador_seis / N

# Imprime o resultado
print("A probabilidade de obter pelo menos um dado com seis ao lançar 5 dados é:", probabilidade_final)

# Gera o histograma
plt.bar(["Pelo menos um seis", "Nenhum seis"], [contador_seis, N - contador_seis], color=['blue', 'red'], alpha=0.7)
plt.title("Probabilidade de obter pelo menos um seis ao lançar 5 dados")
plt.xlabel("Resultado")
plt.ylabel("Frequência")
plt.show()

# Gera o gráfico da probabilidade
plt.plot(range(1, N + 1), probabilidades, color='green')
plt.title("Probabilidade de obter pelo menos um seis ao lançar 5 dados")
plt.xlabel("Número de experimentos")
plt.ylabel("Probabilidade")
plt.show()
```
