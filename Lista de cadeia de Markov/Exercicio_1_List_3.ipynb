```python
import numpy as np
from numpy.linalg import matrix_power
import matplotlib.pyplot as plt

##########Analise sumulado################
n = 100000
Value1=np.zeros(n, int)

for j in np.arange(0, n, 1):
    probabilidades_transicao = [[0,1/2, 1/2 , 0,0,0],[1/2,0,0,1/2,0,0],[1/3,0,0,1/3,1/3,0],[0,1/3,1/3, 0 ,0,1/3], [0,0,1/2, 0,0 ,1/2] , [0,0,0,0,0,1]]
    valor_inicial = 0  # 0 estado 0,  1 estado 1,  2 estado 2, 3 estado 3, 4 estado 4 , 5 estado 5
    chain_length = 11
    chain = np.zeros(chain_length, int)
    chain[0] = valor_inicial
    for i in np.arange(1, chain_length, 1):
        Linha_atual = probabilidades_transicao[chain[i - 1]]
        #print(Linha_atual)
        acumulada = np.cumsum(Linha_atual)
        r = np.random.uniform(0, 1)
        chain[i] = np.argmax(acumulada > r)
    Value1[j] = chain[3]

prob1 = np.sum(Value1 == 5) / n

print("Resultados simulado")

print("A probabilidade do rato morrer apoes 3 horas (Evento simulado) é  " , prob1)


#####################Analise analitico ##############################

probabilidades_transicao = [[0,1/2, 1/2 , 0,0,0],[1/2,0,0,1/2,0,0],[1/3,0,0,1/3,1/3,0],[0,1/3,1/3, 0 ,0,1/3], [0,0,1/2, 0,0 ,1/2] , [0,0,0,0,0,1]]

xn3=matrix_power(probabilidades_transicao, 3)
print("Resultados Analiticos")

print("A probabilidade do rato morrer apoes 3 horas (Evento analitico) é " ,xn3[0][5])

##################### Plot ##############################

probabilidades_transicao = [[0,1/2, 1/2 , 0,0,0],[1/2,0,0,1/2,0,0],[1/3,0,0,1/3,1/3,0],[0,1/3,1/3, 0 ,0,1/3], [0,0,1/2, 0,0 ,1/2] , [0,0,0,0,0,1]]

valor_inicial = 0  # 0 estado 0,  1 estado 1,  2 estado 2, 3 estado 3, 4 estado 4 , 5 estado 5
chain_length = 30
chain = np.zeros(chain_length, int)
chain[0] = valor_inicial
for i in np.arange(1, chain_length, 1):
    Linha_atual = probabilidades_transicao[chain[i - 1]]
    #print(Linha_atual)
    acumulada = np.cumsum(Linha_atual)
    r = np.random.uniform(0, 1)
    chain[i] = np.argmax(acumulada > r)

i=np.arange(0, chain_length, 1)
plt.plot(i, chain[i],'o-')
plt.xlabel('Numero de passo')
plt.ylabel('Estado de transição ')
plt.show()
plt.show()
```
