## Geração do Sinal 4-PAM

Este código, gera um sinal 4-PAM onde, os símbolos são gerados aleatoriamente a partir dos níveis possíveis (-3, -1, 1, 3) e armazenados na variável "simbolos_transmitidos".

O gráfico resultante mostra como os símbolos são organizados ao longo de um eixo temporal, representando a sequência de símbolos transmitida.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros para a geração do sinal
num_simbolos = 1000  # Número de símbolos
ordem_pam = 4  # Ordem PAM
niveis_simbolos = np.arange(-(ordem_pam - 1), ordem_pam, 2)

# Gera símbolos aleatórios para o sinal 4-PAM
simbolos_transmitidos = np.random.choice(niveis_simbolos, num_simbolos)

# Imprime os primeiros 10 valores transmitidos
print("Primeiros 10 símbolos transmitidos:", simbolos_transmitidos[:10])

# Plota o sinal 4-PAM transmitido
plt.figure(figsize=(10, 4))
plt.plot(simbolos_transmitidos, drawstyle='steps-mid')
plt.title('Sinal 4-PAM Transmitido')
plt.xlabel('Índice do Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

## Demodulação do Sinal 4-PAM após um Canal Ruidoso:

Aqui, Um canal ruidoso é simulado ao adicionarmos ruído Gaussiano branco com desvio padrão (desvio_padrao_ruido) ao sinal transmitido.
A variável simbolos_recebidos contém os símbolos transmitidos mais o ruído.

Durante a demodulação, cada símbolo recebido é comparado aos níveis de símbolo originais, e atribui-se o valor mais próximo (funciona como uma quantização).

O gráfico mostra os símbolos recebidos, seguidos pelo gráfico do sinal demodulado.


```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Gera um canal ruidoso (AWGN)
num_simbolos = 1000  # Número de símbolos
ordem_pam = 4  # Ordem PAM
niveis_simbolos = np.arange(-(ordem_pam - 1), ordem_pam, 2)
simbolos_transmitidos = np.random.choice(niveis_simbolos, num_simbolos)
desvio_padrao_ruido = 0.5  # Desvio padrão do ruído
simbolos_recebidos = simbolos_transmitidos + np.random.normal(0, desvio_padrao_ruido, num_simbolos)

# Imprime os primeiros 10 valores recebidos
print("Primeiros 10 símbolos recebidos (com ruído):", simbolos_recebidos[:10])

# Demodula os símbolos recebidos
simbolos_demodulados = np.zeros(num_simbolos)
for i in range(num_simbolos):
    simbolos_demodulados[i] = min(niveis_simbolos, key=lambda x: abs(simbolos_recebidos[i] - x))

# Imprime os primeiros 10 valores demodulados
print("Primeiros 10 símbolos demodulados:", simbolos_demodulados[:10])

# Plota o sinal recebido (com ruído)
plt.figure(figsize=(10, 4))
plt.plot(simbolos_recebidos, drawstyle='steps-mid', label='Sinal 4-PAM Recebido com Ruído')
plt.xlabel('Índice do Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.title('Sinal 4-PAM Recebido com Ruído')
plt.show()

# Plota o sinal demodulado
plt.figure(figsize=(10, 4))
plt.plot(simbolos_demodulados, drawstyle='steps-mid', label='Sinal 4-PAM Demodulado', color='green')
plt.xlabel('Índice do Símbolo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.title('Sinal 4-PAM Demodulado')
plt.show()
```
